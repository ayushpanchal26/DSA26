```python
def my_brutforce(arr,target):
    n = len(arr)
    for i in range(n):
        # for j in range(n):f
        for j in range(1,n):
            if arr[i] + arr[j] == target:
                return i,j
        
arr = [2,6,5,8,11]
target = 14
print(my_brutforce(arr,target))
```

```python
def hashing(arr,target):
    n = len(arr)
    mapp = {}
    for i in range(n):
        num  = arr[i]
        more_needed = target - num
        if more_needed in mapp:
            return [mapp[more_needed],i]
        
        mapp[num] = i
    return -1

arr = [2,6,5,8,11]
target = 14
print(hashing(arr,target))
        
```

```python
def two_pointer_greedy(arr,target):
    arr.sort()
    left = 0
    right = len(arr)-1

    while left<right:
        sum = arr[left] + arr[right]

        if sum == target:
            return True
        
        elif sum < target:
            left +=1 
        else:
            right -= 1

    return -1

arr = [2,6,5,8,11]
target = 14
print(two_pointer_greedy(arr,target))
```

Majority Element (>n/2 times)

```python
def my_burtforce(arr,n):

    for i in range(n):
        count = 0
        for j in range(n):
            if arr[i] == arr[j]:
                count +=1
        
        if count > n//2:
            return arr[i]
        
    return -1


arr = [2,2,1,1,1,2,2]
my_burtforce(arr,len(arr))
```

```python
# advance python
from collections import Counter
def better(arr):
    n = len(arr)

    counter = Counter(arr)

    for num, count in counter.items():
        if count> (n//2):
            return num
        
    return -1
# O(N*logN) + O(N)  
arr = [2,2,1,1,1,2,2]
print(better(arr))
```

```python
def brutforce(arr):
    n = len(arr)
    pos = []
    neg = []

    for i in range(n):
        if arr[i] >0:
            pos.append(arr[i])
        else:
            neg.append(arr[i])

    
    for i in range(len(pos)):
        arr[2*i] = pos[i]
    for i in range(len(neg)):
        arr[2*i+1] = neg[i]

    return arr

if __name__ == "__main__":
    arr = [1,2,-4,-5]
    ans = brutforce(arr)

    for i in range(len(ans)):
        print(ans[i],end=" ")
```

```python
def brutforce(arr):
    n = len(arr)
    temp = []
    for i in range(n):
        leader = True
        for j in range(i+1,n):
            if arr[j] > arr[i]:
                leader = False
                break
        
        if leader:
            temp.append(arr[i])

    return temp

if __name__ == "__main__":
    arr = [10,22,12,3,0,6]
    ans = brutforce(arr)
    print(ans)
```

longest consecutive sequence 

```python
def linear(arr,x):
    n = len(arr)
    for i in range(n):
        if arr[i] == x:
            return True
    return False

def brutforce(arr):
    n = len(arr)
    longest = 1
    for i in range(n):
        cnt = 1
        x = arr[i]

        while linear(arr,x+1):
            x+=1
            cnt+=1

        longest = max(cnt,longest)

    return longest

arr = [100,200,1,2,3,4]
ans = brutforce(arr)
print("The longest consecutive sequence is",ans)
```
